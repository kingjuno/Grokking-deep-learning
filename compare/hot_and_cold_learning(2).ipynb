```python
weight = 0.5
input = 0.5
goal_prediction = 0.8

step_amount = 0.001
```

```python
for iteration in range(1101):
    prediction = input*weight
    error = (prediction-goal_prediction)**2
    print(f'{iteration}: Error: {error} Prediction: {prediction}')
    prediction_up = input*(weight+step_amount)
    prediction_down = input*(weight-step_amount)

    error_up = (prediction_up-goal_prediction)**2
    error_down = (prediction_down-goal_prediction)**2

    if(error_up>error_down):
        weight-=step_amount
    else:
        weight+=step_amount
```
